```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from sklearn.metrics import accuracy_score

le = LabelEncoder()
# Read the csv file and store it in df
df = pd.read_csv('/content/expt6file.csv')
# Encodes string to integer
df1 = df.apply(le.fit_transform)

# analysis for selection of input features using seaborn's heatmap
data1 = pd.DataFrame(df1)
corr = data1.corr()
print(corr['Play?'].sort_values(ascending=False), "\n")
sns.heatmap(data1.corr(), annot=True)
plt.show()

X = df1.iloc[:, [1, 4]].values
Y = df1.iloc[:, 5].values
# Divides the data to be used to train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=1, stratify=df1['Play?'])

# Cluster Analysis
KMeans_Clustering = KMeans(n_clusters=2, random_state=0)
KMeans_Clustering.fit(X_train, Y_train)
print("Centroids of 2 clusters are:\n", KMeans_Clustering.cluster_centers_)
print("Clusters:\n", KMeans_Clustering.labels_)
pred = KMeans_Clustering.predict(X_test)
print("Prediction:", pred)
print("Accuracy:", accuracy_score(Y_test, pred))

toplay = KMeans_Clustering.predict([[2, 0]])
print("Play?:", toplay)

# Cluster
plt.scatter(df1.iloc[:, [1]].values, df1.iloc[:, [4]].values, c=df1['Play?'], cmap='rainbow')
plt.xlabel(f"{df1['Weather'].name}")
plt.ylabel(f"{df1['Wind'].name}")
plt.show()

# Predicted Cluster
kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(X)
plt.scatter(df1.iloc[:, [1]].values, df1.iloc[:, [4]].values, c=kmeans.labels_, cmap='rainbow')
plt.xlabel('Weather')
plt.ylabel('Wind')
plt.show()
```

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn.metrics import accuracy_score

# Read the csv file and store it in df
df = pd.read_csv('/content/data.csv')

# analysis for selecting the input features for training the model
data = pd.DataFrame(df)

# calculating the correlation of the data
corr = data.corr()
print(corr['Outcome'].sort_values(ascending=False), "\n")
print(sns.heatmap(data.corr(), annot=True))

x = df.iloc[:, [1, 5, 7]]
y = df.iloc[:, [8]]

# Divides the data to be used to train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1, stratify=df['Outcome'])

# training the model
kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(x_train)

# printing the centroids and the actual clusters
print("Centroids of 2 clusters are:\n", kmeans.cluster_centers_)
print("Clusters:\n", kmeans.labels_)
pred = kmeans.predict(x_test)
print("Predict:\n", pred)
print("Accuracy:\n", accuracy_score(y_test, pred))
# actual cluster output
plt.scatter(df.iloc[:, [1]].values, df.iloc[:, [5]].values, c=df['Outcome'], cmap='rainbow')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()

# predicted cluster
kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(x)
plt.scatter(df.iloc[:, [1]].values, df.iloc[:, [5]].values, c=kmeans.labels_, cmap='rainbow')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```
